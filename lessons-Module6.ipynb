```python
# Sum of n numbers squared

n1 = [1, 4, 6, 10, 20]
n2 = [0.02, 0.04 , 1.6 , 1.1]
n3 = [1, 3, 4, 5, 6, 9, 10, 19, 20]
n4 = [1, 3, 4, 5, 6, 9, 10, 19, 20, 20, 20, 20, 20]



def sum_of_squared_evens(numbers:list):
    ''' sum of the squares of only the even numbers
    
    args:
        numbers (list): list of numbers
    
    return:
        (float): sum of the even numbers squared
    '''
    if len(numbers) > 10:
        s = 'too much work for me'
        s = 0
        return s

    our_sum = 0
    for n in numbers:
        if n % 2 == 0:
            # this is even
            # our_sum = our_sum + n**2
            our_sum += n**2

    return our_sum

# outside function (back in the global scope)

result_a = sum_of_squared_evens(n1)
result_b = sum_of_squared_evens(n2)
result_c = sum_of_squared_evens(n3)
result_d = sum_of_squared_evens(n4)

print('the sum of the even squares:', result_a)
print('the sum of the even squares:', result_b)
print('the sum of the even squares:', result_c)
print('the sum of the even squares:', result_d)

# make a master list
all_lists = [n1, n2, n3, n4]

print('\nthe sum of the even squares:')
# total = 0
for x in all_lists:
    # total = total + sum_of_squared_evens(x)
    print(sum_of_squared_evens(x))

# print('the total is', total)

# When do we only write return (without a value)
```

```python
class Dog:

    # shared with all the instances
    kind = 'canine'
    food_types = []    # <-- common to all  (power of a community)
    
    # unique to each instance (constructor, initialise)
    def __init__(self, name, number_of_legs=4, 
                            its_first_trick='jump and spin') -> None:
        self.name = name
        self.tricks = [its_first_trick]
        self.number_of_legs = number_of_legs


    def do_a_trick(self, our_trick:str):
        ''' teach the dog do a trick 
        
        args:
            our_trick (str): a trick that the dog can learn

        appends the list of tricks.
        '''
        self.tricks.append(our_trick)
    

    def likes_food(self, my_food:str):
        ''' introduce a new food type 
        
        args:
            my_food (str): some food that the dog likes

        appends the food list.        
        '''
        self.food_types.append(my_food)

print('everythig about dogs is created here...')
```

```python
class Dog:

    # shared with all the instances
    kind = 'canine'
    food_types = []    # <-- common to all
    
    # unique to each instance
    def __init__(self, name, number_of_legs=4) -> None:
        self.name = name
        self.tricks = []
        self.number_of_legs = number_of_legs


    def do_a_trick(self, our_trick:str):
        ''' teach the dog do a trick '''
        self.tricks.append(our_trick)

    def likes_food(self, my_food):
        ''' introduce a new food type '''
        self.food_types.append(my_food)

    def __repr__(self) -> str:
        ''' user friendly output string '''
        s= '<class Dog>\n'
        s = s + f'name: {self.name}, '
        s = s + f'tricks: {self.tricks}'
        return s


# create some dog instances
d = Dog('scruffy')
e = Dog('bob')
f = Dog('fred', number_of_legs=3)


print(d.__dict__)
print('---')
print(d)

```

```python
'''
example of classes

'''

class RegularPerson:
    ''' description of the class '''

    def __init__(self, age, height):
        self.age = age
        self.height = height
    
    def increase_age(self):
        self.age = self.age + 1

# class child(parent)    # child inherits from the parent

class AnnoyingPerson(RegularPerson):
    ''' description of the class '''

    def __init__(self, age, height, picks_nose, chews_loudly):
        self.picks_nose = picks_nose
        self.chews_loudly = chews_loudly
        RegularPerson.__init__(self, age, height)
         # you can also use super().__init__(age, height)

    def really_annoying(self):
        if self.picks_nose is not None and self.chews_loudly is not None:
            print('very annoying')
        else:
            print('not so bad')
    
    def increase_level(self):
        ''' make one attribute more annoying '''

        if self.picks_nose is not None:
            self.picks_nose = self.picks_nose + 1
        else:
            self.picks_nose = 1


    def smelly_feet_checker(self):
        ''' check for the smelly feet attribute '''

        if hasattr(self, 'smelly_feet'):
            print('they have smelly feet')
        else:
            print('they have clean feet')


# class Child(Parent):
class DynamicPerson(AnnoyingPerson):

    def __init__(self,annoy_person:AnnoyingPerson, plays_chess:bool , runs_fast:bool):
        self.plays_chess = plays_chess
        self.runs_fast = runs_fast
        self.annoy_person = annoy_person
        AnnoyingPerson.__init__(self, annoy_person.age, annoy_person.height, 
            annoy_person.picks_nose, annoy_person.chews_loudly)

    def features_good_bad(self):
        if self.plays_chess and self.age:
            print(self.age)
        
        

person1 = AnnoyingPerson(20, 1.5, 100, 100)
person2 = AnnoyingPerson(15, 1.1, None, 100)
person3 = AnnoyingPerson(50, 1.1, 100, None)

person1.age = 10
person2.age = 20
person3.age = 20

person1.really_annoying()
person2.really_annoying()
person3.really_annoying()


person2.increase_level()
print(person2.picks_nose)

person2.really_annoying()

person2.age = 10
print(person2.__dict__)

person1.increase_age()
person2.increase_age()

print(person1.__dict__)
print(person2.__dict__)

person2.smelly_feet = True
print(person2.__dict__)

person2.smelly_feet_checker()
person1.smelly_feet_checker()


personA = DynamicPerson(person1, True,True)
personA.features_good_bad()





```

```python
class Warehouse:
    # shared attributes
    purpose = 'storage'
    region = 'west'
    object_in_warehouse = ['table', 'chair', 'sofa']

    def __init__(self) -> None:
        self.object_in_warehouse=[]

    


# created instances. instantiate 
w1 = Warehouse
w2 = Warehouse

# so lets try this.... more desired behaviour
w1 = Warehouse()
w2 = Warehouse()


# if we change the shared attribute of one, then the other one changes too
w2.region = 'south'

# here is the proof
print(w1.__dict__)
print(w2.__dict__)

print(w1.region, w2.region)


w1.object_in_warehouse.append('bed')
print(w1.object_in_warehouse)
print(w2.object_in_warehouse)
print('---')
print(w1.__dict__)
print(w2.__dict__)
```
